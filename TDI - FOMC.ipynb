```python
import nltk
import nltk.tokenize
import nltk.stem
import nltk.collocations
import nltk.tag
import nltk.parse
import nltk.probability
import numpy
import PyPDF2
from nltk.corpus import stopwords
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path=r'C:\Users\Niloo\Desktop\FRE\Spring 18\News\HW1\FOMC Minutes\FOMC Minutes text\\'
os.listdir(path)
```

```python
n_emp=[]
n_reces=[]
for file in os.listdir(path):
    f = open (path+file)
    raw=f.read()
    tokens = nltk.word_tokenize(raw)
    text = nltk.Text(tokens)
    emp=text.count("employment")
    Emp=text.count("Employment")
    reces=text.count("recession")
    Reces=text.count("Recession")
    number_emp=emp+Emp
    number_rec=Reces+reces
    n_emp.append(number_emp)
    n_reces.append(number_rec)
n_emp
n_reces
```

```python
x=[1,3,5,6,7,9,10,12,]
plt.plot(x,n_emp,'-o',label="employment")
plt.plot(x,n_reces,'-ok',label="recession")
plt.legend()
plt.ylabel('#')
plt.xlabel('months')
plt.show()
```
